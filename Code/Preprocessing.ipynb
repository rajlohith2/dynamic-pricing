```python
import modin.pandas as mpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
```

```python
train_data = pd.read_excel('train.xlsx')
train_data.head()
print(len(train_data['id']))
```

```python
cust_id = list(train_data['id'])
print(cust_id[:5])
```

```python
chunk_size = 2 * (10 ** 7)
df = mpd.DataFrame()
count=0
for chunk in mpd.read_csv('transactions.csv.gz',compression='gzip',engine = 'c',chunksize = chunk_size):
    print(chunk.shape)
    filename = "num_"+str(count)+".csv"
    count+=1
    chunk.loc[chunk['id'].isin(cust_id)].to_csv(filename)
    
```

```python
num0_df = pd.read_csv("num_0.csv")
num1_df = pd.read_csv("num_1.csv")
num2_df = pd.read_csv("num_2.csv")
num3_df = pd.read_csv("num_3.csv")
num4_df = pd.read_csv("num_4.csv")
num5_df = pd.read_csv("num_5.csv")
num6_df = pd.read_csv("num_6.csv")
num7_df = pd.read_csv("num_7.csv")
num8_df = pd.read_csv("num_8.csv")
num9_df = pd.read_csv("num_9.csv")
num10_df = pd.read_csv("num_10.csv")
num11_df = pd.read_csv("num_11.csv")
num12_df = pd.read_csv("num_12.csv")
num13_df = pd.read_csv("num_13.csv")
num14_df = pd.read_csv("num_14.csv")
num15_df = pd.read_csv("num_15.csv")
num16_df = pd.read_csv("num_16.csv")
num17_df = pd.read_csv("num_17.csv")
```

```python
hash_map_purchase_val = {}
num_df_arr = [num0_df,num1_df,num2_df,num3_df,num4_df,num5_df,num6_df,num7_df,num8_df,num9_df,num10_df,num11_df,num12_df,num13_df,num14_df,num15_df,num16_df,num17_df]

for num_df in num_df_arr:
    for ind in num_df.index:
        if(num_df['id'][ind] in cust_id):
            if(num_df['id'][ind] not in hash_map_purchase_val):
                hash_map_purchase_val[num_df['id'][ind]] = 0
            hash_map_purchase_val[num_df['id'][ind]] += num_df['purchaseamount'][ind]

    
```

```python
print(hash_map_purchase_val)
```

```python
hash_map_purchase_cat = {}
for num_df in num_df_arr:
    for ind in num_df.index:
        if(num_df['id'][ind] in cust_id):
            if(num_df['id'][ind] not in hash_map_purchase_cat):
                hash_map_purchase_cat[num_df['id'][ind]] = []
            if(num_df['category'][ind] not in hash_map_purchase_cat[num0_df['id'][ind]]):
                hash_map_purchase_cat[num_df['id'][ind]].append(num0_df['category'][ind])
```

```python
print(hash_map_purchase_cat)

    
    
```

```python
hash_map_purchase_cat_len = {}
for key,val in hash_map_purchase_cat.items():
    hash_map_purchase_cat_len[key] = len(val)
print(hash_map_purchase_cat_len)
```

```python
hash_map_purchase_brand = {}
for num_df in num_df_arr:
    for ind in num_df.index:
        if(num_df['id'][ind] in cust_id):
            if(num_df['id'][ind] not in hash_map_purchase_brand):
                hash_map_purchase_brand[num_df['id'][ind]] = []
            if(num_df['brand'][ind] not in hash_map_purchase_brand[num_df['id'][ind]]):
                hash_map_purchase_brand[num_df['id'][ind]].append(num_df['brand'][ind])
```

```python
hash_map_purchase_brand_len = {}
for key,val in hash_map_purchase_brand.items():
    hash_map_purchase_brand_len[key] = len(val)
print(hash_map_purchase_brand_len)
```

```python
hash_map_purchase_quantity = {}
for num_df in num_df_arr:
    for ind in num_df.index:
        if(num_df['id'][ind] in cust_id):
            if(num_df['id'][ind] not in hash_map_purchase_quantity):
                hash_map_purchase_quantity[num_df['id'][ind]] = 0
            hash_map_purchase_quantity[num_df['id'][ind]]+= num_df['purchasequantity'][ind]
```

```python
hash_map_purchase_company = {}
for num_df in num_df_arr:
    for ind in num_df.index:
        if(num_df['id'][ind] in cust_id):
            if(num_df['id'][ind] not in hash_map_purchase_company):
                hash_map_purchase_company[num_df['id'][ind]] = []
            if(num_df['company'][ind] not in hash_map_purchase_company[num_df['id'][ind]]):
                hash_map_purchase_company[num_df['id'][ind]].append(num_df['company'][ind])
```

```python
hash_map_purchase_company_len = {}
for key,val in hash_map_purchase_company.items():
    hash_map_purchase_company_len[key] = len(val)
print(hash_map_purchase_company_len)
```

```python
customer_data = pd.read_excel("train.xlsx")
```

```python
customer_data.head()
```

```python
for ind in customer_data.index:
    try:
        customer_data['PCT'][ind] = round(round(hash_map_purchase_val[customer_data['id'][ind]],3)/hash_map_purchase_cat_len[customer_data['id'][ind]],3)
        customer_data['PQT'][ind] = round(round(hash_map_purchase_val[customer_data['id'][ind]],3)/hash_map_purchase_quantity[customer_data['id'][ind]],3)
        customer_data['PCY'][ind] = round(round(hash_map_purchase_val[customer_data['id'][ind]],3)/hash_map_purchase_company_len[customer_data['id'][ind]],3)
        customer_data['PBD'][ind] = round(round(hash_map_purchase_val[customer_data['id'][ind]],3)/hash_map_purchase_brand_len[customer_data['id'][ind]],3)
        customer_data['Purchase_amount'][ind] = round(hash_map_purchase_val[customer_data['id'][ind]],3)
        customer_data['Purchase_quantity'][ind] = hash_map_purchase_quantity[customer_data['id'][ind]]
    except:
        print(customer_data['id'][ind])
        pass

customer_data.to_csv("final_processed.csv")
```

```python
processed_csv = pd.read_csv("final_processed.csv")
processed_csv.drop('Unnamed: 0.1',axis=1)

```

```python
offer_data = pd.read_csv("offers.csv.gz",compression='gzip')
print(offer_data.head())
offer_data_hash = {}
for ind in offer_data.index:
    if(offer_data['offer'][ind] not in offer_data_hash):
        offer_data_hash[offer_data['offer'][ind]] = offer_data['offervalue'][ind]
print(offer_data['offer'].value_counts())
```
